```python
comments_and_videos = pd.read_excel('comments_with_predictions.xlsx')
```

```python
comments_and_videos.head()
```

```python
transcripts = transcripts_with_titles.rename(columns={
    'Sentiment': 'video_sentiment',
    'Topic_Name': 'video_topic',
    'Emotion': 'video_emotion',
    'Hate_Speech_Label': 'video_hate_speech_label'
})
```

```python
transcripts.head()
```

```python
comments_and_videos = comments_and_videos.merge(transcripts[['title', 'video_sentiment', 'video_topic', 'video_emotion', 'video_hate_speech_label']],
                             on='title', how='left')
```

```python
comments_and_videos.head()
```

```python
import matplotlib.pyplot as plt
```

1. Hate speech Distribution in Comments

```python

sentiment_counts = comments_and_videos['Predicted_Label'].value_counts()
plt.figure(figsize=(8, 6))
sentiment_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=["#66c2a5", "#fc8d62", "#8da0cb"])
plt.title("Sentiment Distribution in Comments")
plt.ylabel("")
plt.show()
```

2. Emotion vs. Hate Speech in Comments

```python
hateful_comments = comments_and_videos[comments_and_videos['Predicted_Label'] == 'insult']

emotion_hate_speech = hateful_comments['video_emotion'].value_counts(normalize=True) * 100

# Plot the data
emotion_hate_speech.plot(kind='bar', color="#fc8d62", figsize=(8, 6))
plt.title("Proportion of Hateful Comments by Video Emotion")
plt.xlabel("Video Emotion")
plt.ylabel("Percentage of Hateful Comments")
plt.xticks(rotation=45)
plt.show()
```

3. Topics and Comment Sentiment

```python
import seaborn as sns

topic_sentiment = comments_and_videos.groupby(['video_topic', 'Predicted_Label']).size().unstack()

topic_sentiment.plot(kind='bar', stacked=True, figsize=(12, 6), colormap="viridis")
plt.title("Sentiment Distribution Across Video Topics")
plt.xlabel("Video Topic")
plt.ylabel("Number of Comments")
plt.xticks(rotation=45)
plt.legend(title="Comment Sentiment")
plt.tight_layout()
plt.show()
```

```python
channel_hate_speech = comments_and_videos[comments_and_videos['Predicted_Label'] == 'insult']['Channel'].value_counts(normalize=True) * 100

# Plot the data
channel_hate_speech.plot(kind='bar', color=["#66c2a5", "#fc8d62"], figsize=(8, 6))
plt.title("Proportion of Hateful Comments by Channel")
plt.xlabel("Channel")
plt.ylabel("Percentage of Hateful Comments")
plt.xticks(rotation=0)
plt.show()
```

```python

comments_filtered = comments_and_videos[comments_and_videos['video_emotion'] != 'joy']

emotion_sentiment_heatmap = comments_filtered.pivot_table(
    index='video_emotion',
    columns='Predicted_Label',
    aggfunc='size',
    fill_value=0
)

plt.figure(figsize=(10, 6))
sns.heatmap(emotion_sentiment_heatmap, annot=True, fmt="d", cmap="coolwarm", cbar=True)
plt.title("Heatmap: Video Emotion (Excluding Joy) vs. Comment Sentiment")
plt.xlabel("Comment Sentiment")
plt.ylabel("Video Emotion")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
topic_hate_percentage.sort_values().plot(kind='bar', color="#fc8d62")
plt.title("Percentage of Hate Speech Comments by Topic")
plt.xlabel("Video Topic")
plt.ylabel("Percentage of Hate Speech Comments")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
comments_filtered = comments_and_videos[comments_and_videos['video_emotion'] != 'joy']
```

```python
hate_speech_stats = comments_filtered.groupby(['video_topic', 'video_emotion']).apply(
    lambda x: (x['Predicted_Label'] == 'insult').mean() * 100
).reset_index(name='hate_speech_percentage')
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=hate_speech_stats,
    x='video_topic',
    y='video_emotion',
    size='hate_speech_percentage',
    hue='hate_speech_percentage',
    palette='coolwarm',
    sizes=(50, 500)
)
plt.title("Hate Speech Percentage by Topic and Emotion (Excluding Joy)")
plt.xlabel("Video Topic")
plt.ylabel("Video Emotion")
plt.xticks(rotation=45)
plt.legend(title="Hate Speech %", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
