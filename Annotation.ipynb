Sample 2000 comments for annotation


```python
if len(data) < 2000:
    print(f"Dataset has only {len(data)} comments, which is fewer than 2000.")
else:
    # Randomly sample 2000 comments
    sample = data.sample(n=2000, random_state=42)

    # Save the sampled data to a new CSV file (optional)
    sample.to_csv('sampled_comments.csv', index=False)
    print("Sampled 2000 comments successfully and saved to 'sampled_comments.csv'.")

# Preview the sample
print(sample.head())
```

```python
from google.colab import files

# Assuming `sample` is the DataFrame containing the sampled data
# Save the sampled dataset to an Excel file
sample.to_excel('sampled_comments.xlsx', index=False)

# Download the file to your local machine
files.download('sampled_comments.xlsx')
```
